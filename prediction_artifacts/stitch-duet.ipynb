```python
# duet_stitch = pd.DataFrame()
duet_stitch = pd.DataFrame(columns=["username", "video_description", "id"])

# for i in range(1, len(videos)):
#     for video in videos[i,:]:
#         for description in video["video_description"]:
#             if "#duet with" in description:
#                 duet_stitch.append(video["username", "video_description", "id"]) 
#             elif "#stitch with" in description:
#                 duet_stitch.append(video["username", "video_description", "id"]) 

for _, video in videos.iterrows():  
    description = video["video_description"]  # Extract description text
    
    if isinstance(description, str):  # Ensure it's a string to avoid errors
        if "#duet with" in description or "#stitch with" in description:
            # Append new row to the DataFrame
            duet_stitch = pd.concat([duet_stitch, pd.DataFrame([{
                "username": video["username"],
                "video_description": video["video_description"],
                "id": video["id"]
            }])], ignore_index=True)

# Display the new DataFrame
print(duet_stitch)
```

```python

import re

def format_description(description):
    if isinstance(description, str):  # Ensure it's a string
        description = description.lower()  # Convert to lowercase
        description = re.sub(r"\s+", "", description)  # Remove all spaces
    return description

for i, video in duet_stitch.iterrows():  
    description = video["video_description"]  # Extract description text
    
    duet_stitch.loc[i,"video_description"] = format_description(description)

print(duet_stitch)
```

```python
creators = pd.read_csv("../shared-folder-gald/data/follow-link.csv", index_col=0)
creators
```

```python
matched = []

for _, row in duet_stitch.iterrows():
    video_desc = row['video_description']
    for creator in creators['target'].unique():
        if creator in video_desc[:38]:
            matched.append({
        'username': row['username'],
        'video_description': video_desc,
        'id': row['id'],
        'creator': creator})
            
result_df = pd.DataFrame(matched)
print(result_df.head())
```

```python
# excluding creators that dueted themselves
matched_uni = []

for _, row in duet_stitch.iterrows():
    video_desc = row['video_description']
    for creator in creators['target'].unique():
        if creator in video_desc[:38] and row['username'] != creator:  # Exclude matching username
            matched_uni.append({
                'username': row['username'],
                'video_description': video_desc,
                'id': row['id'],
                'creator': creator
            })

unique_df = pd.DataFrame(matched_uni)
print(unique_df.head())
```

```python
unique_df.to_csv('duet_stitch_uniques.csv', header=True) 
```
